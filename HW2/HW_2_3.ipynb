Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

```python
average_rating_by_author = df.groupby('author').agg({'user_rating': 'mean'})
average_rating_by_author
```

У якого автора середній рейтинг мінімальний? Відповідь: Donna Tartt

```python
author_with_min_avg_rating = average_rating_by_author['user_rating'].idxmin()
author_with_min_avg_rating
```

Який у цього автора середній рейтинг? Відповідь: 3.9

```python
min_avg_rating = average_rating_by_author.loc[author_with_min_avg_rating, 'user_rating']
min_avg_rating
```

З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

```python
num_books_by_author = df.groupby('author').agg({'name': 'count'})
result_df = pd.concat([num_books_by_author, average_rating_by_author], axis=1)
result_df
```

Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)
Який автор перший у списку? Muriel Barbery

```python
sorted_df = result_df.sort_values(by=['name', 'user_rating'])
sorted_df
```
