Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df = pd.DataFrame(data)
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
missing_percentage = df.isnull().mean()*100
missing_percentage
```

Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
df_cleaned = df.dropna(subset=['Язык.программирования'])
df_cleaned = df_cleaned.dropna(axis=1, how='any')
df_cleaned
```

Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped_data = df.groupby('Должность')
print(grouped_data)
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
aggregated_data = grouped_data.agg({'Зарплата.в.месяц':['min', 'max']})
aggregated_data
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg

```python
def fill_avg_salary(group):
    avg_salary = group['Зарплата.в.месяц'].mean()
    group['avg_salary'] = avg_salary
    return group

df_filling_avg = df.groupby(['Язык.программирования', 'Должность']).apply(fill_avg_salary)
df_filling_avg
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
describe_stats = df_filling_avg['avg_salary'].describe()
describe_stats
```

Збережіть отриману таблицю в CSV файл

```python
df_filling_avg.to_csv('df_filling_avg.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Общий.опыт.работы', y='Изменение.зарплаты.за.12.месяцев', data=data)
plt.xlabel('Общий опыт работы (лет)')
plt.ylabel('Изменение зарплаты за 12 месяцев')
plt.title('Зависимость изменения зарплаты от опыта работы')
plt.show()
```

```python
sns.histplot(data=data, x='Зарплата.в.месяц', kde=True)
plt.xlabel('Зарплата в месяц')
plt.ylabel('Частота')
plt.title('Распределение зарплат в месяц')
plt.show()
```

```python
sns.boxplot(x='Город', y='Зарплата.в.месяц', data=data)
plt.xlabel('Город')
plt.ylabel('Зарплата в месяц')
plt.title('Распределение зарплат по городам')
plt.xticks(rotation=90)
plt.show()
```
